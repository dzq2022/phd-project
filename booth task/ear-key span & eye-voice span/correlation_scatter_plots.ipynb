```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file from the Colab content directory
df = pd.read_excel('/content/correlation_scatter_plots.xlsx')

# Filter the data for cycle 2 and 3
df_cycle2 = df[df['cycle'] == 2]
df_cycle3 = df[df['cycle'] == 3]

# Create the plots
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

# Common axis labels and limits
x_label = 'Ear-Key Span'
y_label = 'Eye-Voice Span'
x_limit = [df['ear_key_span'].min(), df['ear_key_span'].max()]
y_limit = [df['eye-voice span'].min(), df['eye-voice span'].max()]

# Plot for Cycle 2
sns.scatterplot(x='ear_key_span', y='eye-voice span', data=df_cycle2, ax=axes[0], color='blue', s=100, alpha=0.7)
sns.regplot(x='ear_key_span', y='eye-voice span', data=df_cycle2, ax=axes[0], color='blue', scatter=False, ci=95)
axes[0].set_title('Scatter Plot for Cycle 2 with 95% Confidence Interval')
axes[0].set_xlabel(x_label)
axes[0].set_ylabel(y_label)
axes[0].set_xlim(x_limit)
axes[0].set_ylim(y_limit)

# Plot for Cycle 3
sns.scatterplot(x='ear_key_span', y='eye-voice span', data=df_cycle3, ax=axes[1], color='green', s=100, alpha=0.7)
sns.regplot(x='ear_key_span', y='eye-voice span', data=df_cycle3, ax=axes[1], color='green', scatter=False, ci=95)
axes[1].set_title('Scatter Plot for Cycle 3 with 95% Confidence Interval')
axes[1].set_xlabel(x_label)
axes[1].set_ylabel(y_label)
axes[1].set_xlim(x_limit)
axes[1].set_ylim(y_limit)

# Save the plot as SVG in the Colab content directory
plt.tight_layout()
plt.savefig('/content/scatter_plots_with_CI.svg')

# Display the plots
plt.show()
```
